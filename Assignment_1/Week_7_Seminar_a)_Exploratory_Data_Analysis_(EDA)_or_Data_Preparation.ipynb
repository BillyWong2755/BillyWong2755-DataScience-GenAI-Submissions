```python
experience_level_mapping = {
    'EN': 1,
    'MI': 2,
    'SE': 3,
    'EX': 4
}
df['experience_level'] = df['experience_level'].replace(experience_level_mapping)

company_size_mapping = {
    'S': 1,
    'M': 2,
    'L': 3
}
df['company_size'] = df['company_size'].replace(company_size_mapping)

print("Experience level and company size columns have been converted.")
display(df[['experience_level', 'company_size']].head())
```

For employment_type, job_title, employee_residence, company_location, apply one-hot encoding, change the columns name into the format of 'column_name+_+items', e.g. 'employment_type_ FT'.

Okay, I will apply one-hot encoding to employment_type, job_title, employee_residence, and company_location columns. The new columns will be named as specified.



```python
columns_to_encode = ['employment_type', 'job_title', 'employee_residence', 'company_location']

for col in columns_to_encode:
    # Apply one-hot encoding
    dummies = pd.get_dummies(df[col], prefix=col)

    # Concatenate the new dummy variables to the original DataFrame
    df = pd.concat([df, dummies], axis=1)

    # Drop the original column
    df = df.drop(columns=[col])

print("One-hot encoding applied to specified columns.")
display(df.head())
```
